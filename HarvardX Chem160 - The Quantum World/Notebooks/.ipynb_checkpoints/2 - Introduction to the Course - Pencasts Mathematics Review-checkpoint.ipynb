# Brief Review of Vector Analysis
[[back to top]](#Table-of-Contents)

**Position vector** 

$\mathbf{r}(t) = x(t)\mathbf{\hat{x}} + y(t)\mathbf{\hat{y}} + z(t)\mathbf{\hat{z}}$ where $|\mathbf{\hat{x}}| = |\mathbf{\hat{y}}| = |\mathbf{\hat{z}}| = 1$.

**Velocity**
\begin{align*} \mathbf{v} &= \frac{d}{dt}\mathbf{r} = \frac{d}{dt}x(t)\mathbf{\hat{x}} + \frac{d}{dt}y(t)\mathbf{\hat{y}} + \frac{d}{dt}z(t)\mathbf{\hat{z}} \\
&= \mathbf{\dot{r}} = \dot{x}(t)\mathbf{\hat{x}} + \dot{y}(t)\mathbf{\hat{y}} + \dot{z}(t)\mathbf{\hat{z}} \end{align*}

**Dot Product**

- Dot product is a scalar.
- $\mathbf{a} \cdot \mathbf{b} = |\mathbf{a}| \; |\mathbf{b}| \; \cos\theta  = a_xb_x + a_yb_y + a_zb_z $
- Where is the second equality is because $\mathbf{\hat{e}}_i \cdot \mathbf{\hat{e}}_j = 0$ for $i \neq j$ and $1$ when $i=j$. 

**Cross Product**

- The cross product $\mathbf{a} \times \mathbf{b}$ is defined as a vector $\mathbf{c}$ that is perpendicular to both $\mathbf{a}$ and $\mathbf{b}$, with a direction given by the right-hand rule and a magnitude equal to the area of the parallelogram that the vectors span.
- $\vec{A} \times \vec{B} = \left( A_y B_z - A_z B_y \right) \hat{x} - \left( A_x B_z - A_z B_x \right) \hat{y} + \left( A_x B_y - A_y B_x \right) \hat{z}$
- $\left| \mathbf{a} \times \mathbf{b} \right|= |\mathbf{a}| \; |\mathbf{b}| \; \sin\theta $

# Vector Calculus
[[back to top]](#Table-of-Contents)

First we introduce the del operator $\nabla$. Formally, this can be written in Cartesian coordinates as: $\vec{\nabla} = \hat{x}\frac{\partial}{\partial x} + \hat{y}\frac{\partial}{\partial y} + \hat{z}\frac{\partial}{\partial z}$.

**Gradient**

If we let $\nabla$ act on a scalar function $f(x, y, z)$, we get back a vector field called the gradient. 

$$ \vec{\nabla} f(x,y,z) = \hat{x}\frac{\partial f}{\partial x} + \hat{y}\frac{\partial f}{\partial y} + \hat{z}\frac{\partial f}{\partial z} $$

In mathematics, the gradient is a multi-variable generalization of the derivative. While a derivative can be defined on functions of a single variable, for functions of several variables, the gradient takes its place. The gradient is a vector-valued function, as opposed to a derivative, which is scalar-valued.

Like the derivative, the gradient represents the slope of the tangent of the graph of the function. More precisely, the gradient points in the direction of the greatest rate of increase of the function, and its magnitude is the slope of the graph in that direction. The components of the gradient in coordinates are the coefficients of the variables in the equation of the tangent space to the graph. This characterizing property of the gradient allows it to be defined independently of a choice of coordinate system, as a vector field whose components in a coordinate system will transform when going from one coordinate system to another.

**Divergence**

In vector calculus, divergence is a vector operator that produces a scalar field giving the quantity of a vector field's source at each point. More technically, the divergence represents the volume density of the outward flux of a vector field from an infinitesimal volume around a given point.

As an example, consider air as it is heated or cooled. The velocity of the air at each point defines a vector field. While air is heated in a region, it expands in all directions, and thus the velocity field points outward from that region. The divergence of the velocity field in that region would thus have a positive value. While the air is cooled and thus contracting, the divergence of the velocity has a negative value.

$$ \vec{\nabla} \cdot \vec{A}(x,y,z) = \frac{\partial A_x}{ \partial x} + \frac{\partial A_y}{ \partial y} + \frac{\partial A_z}{ \partial z} $$

**Curl**

In vector calculus, the curl is a vector operator that describes the infinitesimal rotation of a vector field in three-dimensional Euclidean space. At every point in the field, the curl of that point is represented by a vector. The attributes of this vector (length and direction) characterize the rotation at that point.

The direction of the curl is the axis of rotation, as determined by the right-hand rule, and the magnitude of the curl is the magnitude of rotation. If the vector field represents the flow velocity of a moving fluid, then the curl is the circulation density of the fluid. A vector field whose curl is zero is called irrotational. The curl is a form of differentiation for vector fields. The corresponding form of the fundamental theorem of calculus is Stokes' theorem, which relates the surface integral of the curl of a vector field to the line integral of the vector field around the boundary curve.

Unlike the gradient and divergence, curl does not generalize as simply to other dimensions; some generalizations are possible, but only in three dimensions is the geometrically defined curl of a vector field again a vector field. This is a similar phenomenon as in the 3 dimensional cross product, and the connection is reflected in the notation for the curl.

$$ \vec{\nabla} \times \vec{A}(x,y,z) = \left( \frac{\partial A_z}{\partial y} - \frac{\partial A_y}{\partial z} \right) \hat{x} - \left( \frac{\partial A_z}{\partial x} - \frac{\partial A_x}{\partial z} \right) \hat{y} + \left( \frac{\partial A_y}{\partial x} - \frac{\partial A_x}{\partial y} \right) \hat{z} $$

# Coordinate Systems
[[back to top]](#Table-of-Contents)

If we select a point, we can describe the location of this point by a position vector, which we call $\vec{r}$. And this position vector $\vec{r}$ can be written as $\vec{r} = x \hat{x} + y \hat{y}$, where the y-component is the projection of the vector on the y-axis, and the x-component is the projection of vector on the x-axis.

In two dimensions, we can convert this to polar coordinates using the projections $x = r \cos(\theta)$ and $y = r \sin(\theta)$. In reverse, we can express $r$ and $\theta$ in terms of $x, y$: $y = r \sin(\theta)$ and $\theta = \arctan \left( \frac{y}{x} \right)$.

Now in three dimensions, we won't use polar coordinates, but we can use something called spherical coordinates. Again, we have our Cartesian axes $x$, $y$, and $z$. If we take a point, we can write it by first projecting onto the xy plane, and then also projecting onto the z-axis.

We can obtain the $x$, $y$, and $z$ components as follows:
\begin{align*} x &= r \sin(\theta) \cos(\phi) \\
y &= r \sin(\theta) \sin(\phi) \\ 
z &= r\cos(\theta) \end{align*}

In spherical coordinates, we introduced two angles. One angle is called phi ($\phi$). We introduce another angle called theta ($\theta$). $\theta$ is the angle that the position vector makes with respect to the z-axis here. $\phi$ is the angle that the position vector makes with respect to the x-axis. Note that in spherical coordinates, $\phi$ takes the same role that $\theta$ had in polar coordinates. Like before, we have $\vec{r} = x \hat{x} + y \hat{y} + z \hat{z}$.

# Complex Numbers
[[back to top]](#Table-of-Contents)

A complex number is usually denoted by the letter $z$. And it has a real part, which we call $x$, and an imaginary part, which is $i$ times $y$, where $i$ is defined as being equal to the square root of minus 1.

\begin{align*} z &= x + iy \\ i &= \sqrt{-1} \end{align*}

It's customary to introduce something called the "complex plane", where we plot the real and imaginary parts in an $xy$ plane. To do this, we define $y$ as equal to the imaginary part of $z$, and $x$ is equal to the real part of $z$: $x = \operatorname{Re}[z]$ and $y = \operatorname{Im}[z]$.

One also defines the complex conjugate, which is written as $z^*$ (say "z star"), with an asterisk. That's obtained by leaving the real part alone, but taking the imaginary part and putting a minus sign in front of it. What it corresponds to in the complex plane is a reflection across the $x$-axis.

There's another form of writing complex numbers. Rather than using Cartesian coordinates, it's sometimes convenient to use a polar form. We can write a complex number using an angle, $\theta$ -- which is the angle with respect to the $x$-axis -- and a radial component, which is the absolute value of the complex number.

In this notation, the absolute value of $z$ -- which is often called $r$, because it's a radial component -- is going to be equal to:

$$ |z| = r = \sqrt{\operatorname{Im}[z]^2 + \operatorname{Re}[z]^2} $$

It's just like the radius of a circle. And $θ$ is going to be

$$ \theta = \arctan\left(\frac{\operatorname{Im}[z]}{\operatorname{Re}[z]}\right) $$

It's just the same as in polar coordinates, it's only the interpretation of $x$ and $y$ and $r$ and $\theta$ are different. We also have that $\operatorname{Re}[z] = r\cos\theta$, and $\operatorname{Im}[z] = r\sin\theta$. This is the same as polar coordinates we have right here.

A way we'll commonly write complex numbers is as an exponential. If we add the real and imaginary components, we have:
$$ z = r\left(\cos\theta + i\sin\theta\right) = re^{i\theta} $$

Another thing that we'll look at are complex functions. There's a whole theory of complex functions called complex analysis, but we'll be actually interested in a very specific case. We'll be interested in complex functions of real variables. The possible states of a quantum system are actually complex functions of real variables.

A complex function of a real variable we can write as:
$$ f(x) = \operatorname{Re} \left[ f(x) \right] + i \operatorname{Im} \left[ f(x) \right] $$

Here, $x$ is a real variable. However, $f$ is a complex function -- it has a real part and an imaginary part.

An example that we'll encounter are the momentum eigenfunctions, $U_p(x)$, which we will introduce when we discuss the momentum operator. These are given by the following expression:

$$ U_p(x) = \frac{1}{\sqrt{2 \pi}} e^{\frac{ipx}{\hbar}} $$

And we can write this in terms of real and imaginary parts, using Euler's formula, in this way:

$$ U_p(x) = \frac{1}{\sqrt{2 \pi}} \left( \cos \left( \frac{px}{\hbar} \right) + i \sin \left( \frac{px}{\hbar}\right)\right) $$

That's a momentum eigenfunction. It's a function of the variable, $x$, which is the position variable. It's a real variable, but the eigenfunction has a real part and an imaginary part. That's an example of a complex function of a real variable.

Basically, we're mapping points from the real line to the complex plane. For instance, the point $x = 0$ would have the location $(0, \frac{1}{\sqrt{2\pi}})$ The location $x = 1$ would be mapped to somewhere to the right, depending on the value of $p$. The location $x$ equals minus 1 would be mapped somewhere below. The point is that the wave function is this type of object. It's a function that takes in real numbers and gives back complex numbers.

# The Dirac Delta Function
[[back to top]](#Table-of-Contents)

In this Pencast, we'll introduce something called Dirac's Delta Function. We use this when we introduce the postulates of quantum mechanics.

Dirac's Delta Function is not actually a function. Technically, mathematicians refer to it as a distribution. And it's denoted in this way: $\delta(x−x^\prime)$.

If we graph it, what it looks like is essentially a spike centered at $x = x^\prime$. Mathematically what we can say is that $\delta(x−x^\prime) = \infty$ if $x$ equals $x^\prime$, and it's equal to $0$ otherwise. If you look at this function, really what it is is a concept rather than a function, because there is no such thing as an infinitely spiked function, but it's a function that's very sharp.

If we imagine taking another function, $f(x)$ let's say, that's smoother -- you can imagine that if you multiply it by the delta function, you're really just going to pick out the value of $f$ of $x$ at $x$ equals $x$ prime.

$$f(x) \ \delta (x - x^\prime) = f(x^\prime) \ \delta (x - x^\prime) $$

This is one property of the delta function, and perhaps the most important: if we take any function and multiply it by $\delta(x−x^\prime)$, we get back a function of the primed variable times the delta function.

The second property of the delta function is that it's normalized to 1. If we integrate the function from minus infinity to infinity, we get 1. The area under the curve is actually a fixed number.

$$ \int_{-\infty}^{\infty} \delta (x - x^\prime) = 1 $$

The third property of the delta function, which actually follows from the first two, is that if we take any function, and we integrate it against the delta function, we get back that function at the point $x^\prime$. Basically again, the function picks out the point. The delta function picks out $f(x)$ only at $x^\prime$.

$$ \int_{-\infty}^{\infty} dx \ f(x) \ \delta (x - x^\prime) = f(x^\prime) $$

Another property, which you shouldn't worry too much about at the moment, but it can be important later on, relates the delta function to a step function.

Let's take a delta function, and instead of integrating it over minus infinity to infinity, we integrate up to some finite value. Let's call it $x^{\prime\prime}$:

$$ \int_{-\infty}^{x^{\prime\prime}} dx \delta (x - x^\prime) = ? $$

What does this mean? Well, if our interval doesn't include $x^\prime$ -- so if $x^{\prime\prime}$ is less than $x^{\prime}$, meaning that I've only integrated up to a lower point -- I get 0. However, suppose I integrate to where $x^{\prime\prime}$ is higher than $x^{\prime}$. Then I've integrated over the whole delta function, and I get 1.

This is often called theta function, or a Heaviside step function.

$$ \int_{-\infty}^{x^{\prime\prime}} dx \ \delta (x - x^{\prime}) = \Theta(x^{\prime\prime}-x^{\prime}) $$
$$ \int_{-\infty}^{x^{\prime\prime}} dx \ \delta (x - x^{\prime}) = 0 \ \text{when} \ x^{\prime\prime} < x^{\prime} \ , \text{or} \ 1 \ \text{when} \ x^{\prime\prime} > x^{\prime} $$

But really, properties one, two, and three are the most important.
